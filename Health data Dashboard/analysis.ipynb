# EDA(Exploratory Data Analysis) + Visualization

```python
df = pd.read_csv('WHO-COVID-19-Dataset-Final.csv')

df['Date_reported'] = pd.to_datetime(df['Date_reported'])

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(x=df['Date_reported'], y=df['Cumulative_cases'], name="Cumulative Cases"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df['Date_reported'], y=df['Cumulative_deaths'], name="Cumulative Deaths"),
    secondary_y=True,
)

fig.update_xaxes(title_text="Date")

fig.update_yaxes(title_text="Cumulative Cases", secondary_y=False)
fig.update_yaxes(title_text="Cumulative Deaths", secondary_y=True)

fig.update_layout(title_text="COVID-19 Cumulative Cases and Deaths Over Time")

fig.show()
```
